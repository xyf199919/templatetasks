```python
import pandas as pd
import pprint
pp = pprint.PrettyPrinter(indent = 4)

xy = pd.read_csv('./samplefinal.csv')

columnlist = xy.columns.tolist()

xy
```

```python
description = xy['Description']
question = [x for x in xy['Question']]
field = [x for x in xy['Answer']]
timing = [x for x in xy['Timing']]
mintime = [x for x in xy['MINIMUM TIME']]
assignment = [x for x in xy['Assignment ']]
client = [x for x in xy['Client Visibility']]
category = [x for x in xy['Category']]


description = [x.replace("- ", "", 1) for x in description]
#need answer onl yes no or maybe
for i, t in enumerate(timing):
    num = int(''.join(filter(str.isdigit, t)))
    startletter= ''.join(filter(lambda x: x.isalpha(), t))[0]
    
    if startletter == 'm':
        timing[i] = num * 30
    elif startletter == 'w':
        timing[i] = num * 7
    elif startletter == 'd':
        timing[i] = num

print(timing)
#need min time
        
for i, c in enumerate(client):
    if c != 'yes':
        client[i] = 'no'
print(client)


# pairs = [[category[i], questions[i]] for i in range(len(category))]
```

```python

  
# Calling DataFrame constructor after zipping 
# both lists, with columns specified 
df = pd.DataFrame(list(zip(description, question, field, timing, mintime, assignment, client, category)), 
               columns =['description', 'question', 'field', 'timing', 'mintime', 'assignment', 'client', 'category']) 
df 


```

```python
content = df.to_csv()
with open('csvfile.csv','wb') as file:
    file.write(content)
    file.close()
```
